```python
import pandas as pd
```

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
# df['koi_disposition']=df['koi_disposition'].map({'CANDIDATE':2,'CONFIRMED':1,'FALSE POSITIVE':0})
# df
y=df.pop('koi_disposition')
from sklearn.preprocessing import LabelEncoder

# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)
```

```python
df
```

```python
encoded_y
```

```python
from sklearn.model_selection import train_test_split # to split the data into two parts
train_x,test_x,train_y,test_y = train_test_split(df, encoded_y,test_size = 0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier # for random forest classifier
model=RandomForestClassifier(n_estimators=100)
model.fit(train_x,train_y)
```

```python
prediction=model.predict(test_x)
```

```python
from sklearn import metrics # for the check the error and accuracy of the model
metrics.accuracy_score(prediction,test_y)
```

```python
# Random Forests in sklearn will automatically calculate feature importance
importances = model.feature_importances_
importances
```

```python
# We can sort the features by their importance
sorted(zip(model.feature_importances_, train_x.columns), reverse=True)
```
